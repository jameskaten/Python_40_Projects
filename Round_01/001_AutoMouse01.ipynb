## Ex2
Automatic weather search

This should open `www.naver.com` in the left window and needs to get exact search type position on the screen

```python
import pyautogui
import time
import pyperclip

pyautogui.moveTo(368, 278, 0.2)  # move to the search position in 0.2s
pyautogui.click()   # mouse click 
time.sleep(0.5)

pyperclip.copy("서울 날씨")     # copy the word
pyautogui.hotkey("ctrl", "v")   # paste it 
time.sleep(0.5)

pyautogui.write(["enter"])
time.sleep(1)
```

 ## Ex4

 Searching many regions and save the data

```python
import pyautogui
import time
import pyperclip

weather = ["서울 날씨", "시흥 날씨", "평촌 날씨", "분당 닐씨", "부산 날씨"]

addr_x = 180
addr_y = 69
start_x = 36
start_y = 307
end_x = 866
end_y = 796

for regions in weather:
    pyautogui.moveTo(addr_x, addr_y, 1)
    time.sleep(0.2)
    pyautogui.click()
    time.sleep(0.2)
    pyautogui.write("www.naver.com", interval=0.1)
    pyautogui.write(["enter"])
    time.sleep(1)
    
    pyperclip.copy(regions)
    pyautogui.hotkey("ctrl", "v")
    time.sleep(0.5)
    pyautogui.write(["enter"])
    time.sleep(1)
    pathss = regions+'.png'
    pyautogui.screenshot(pathss, region=(start_x, start_y, end_x-start_x, end_y-start_y))
    
```
